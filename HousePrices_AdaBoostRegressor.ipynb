```python
base_teste = base_completa.loc[base_completa['SalePrice'].isnull(), :]
base_teste
```

```python
X_base_treino = base_treino.drop(['SalePrice'], axis=1)
y_base_treino = base_treino['SalePrice']
```

```python
X_base_teste = base_teste.drop(['SalePrice'], axis=1)
y_base_teste = base_completa['SalePrice']
```

```python
X_base_treino.shape, y_base_treino.shape
```

```python
X_base_teste.shape, y_base_teste.shape
```

```python
colunas_texto = []
for column in X_base_treino:
    if X_base_treino[column].dtype == 'object':
        colunas_texto.append(column)
colunas_texto
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
for column in colunas_texto:
    X_base_treino[column] = label_encoder.fit_transform(X_base_treino[column])
    X_base_teste[column] = label_encoder.fit_transform(X_base_teste[column])
```

```python
X_base_treino.head()
```

```python
X_base_teste.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_base_treino = scaler.fit_transform(X_base_treino)
X_base_teste = scaler.fit_transform(X_base_teste)
```

```python
X_base_treino, X_base_teste
```

```python
from sklearn.ensemble import AdaBoostRegressor
ada_boost = AdaBoostRegressor()
ada_boost.fit(X_base_treino, y_base_treino)
```

```python
ada_boost.score(X_base_treino, y_base_treino)
```

```python
np.isnan(X_base_teste).any()
```

```python
X_base_teste = np.nan_to_num(X_base_teste)
np.isnan(X_base_teste).any()
```
